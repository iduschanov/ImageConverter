```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = plt.imread('image.jpg')
plt.imshow(img)
```

### Image to grayscale 

Y' = 0.2989 * R' + 0.5870 * G' + 0.1140 * B'

```python
r, g, b = img[:, :, 0], img[:, :, 1], img[:, :, 2]
```

```python
def rgb_to_gray(r, g, b):
    return 0.2989 * r + 0.5870 * g + 0.1140 * b
```

```python
gray = rgb_to_gray(r, g, b)
plt.imshow(gray, cmap=plt.get_cmap('gray'))
```

```python
plt.hist(gray.flatten(), bins=100, color='gray')
plt.show()
```

```python
cv2.imwrite("grayscale.jpg", gray)
```

### Image to YUV

```python
def rgb_to_yuv(img):
     
    x = np.array([[ 0.29900, -0.16874,  0.50000],
                 [0.58700, -0.33126, -0.41869],
                 [ 0.11400, 0.50000, -0.08131]])
     
    yuv = np.dot(img,x)
    yuv[:,:,1:]+=128.0
    return yuv
```

```python
yuv = rgb_to_yuv(img)
```

```python
plt.hist(yuv.flatten(), bins=100)
plt.show()
```

```python
cv2.imwrite("yuv.jpg", yuv)
```
